```python
class Produto:
    def __init__(self, nome, quantidade):
        self.nome = nome
        self.quantidade = quantidade

    def adicionar_estoque(self, quantidade):
        self.quantidade += quantidade

    def remover_estoque(self, quantidade):
        if self.quantidade >= quantidade:
            self.quantidade -= quantidade
        else:
            print(f"Quantidade insuficiente de {self.nome} em estoque.")

    def __str__(self):
        return f"{self.nome}: {self.quantidade} unidades"


class ControleEstoque:
    def __init__(self):
        self.produtos = {}

    def adicionar_produto(self, produto, quantidade):
        if produto.nome not in self.produtos:
            self.produtos[produto.nome] = produto
        self.produtos[produto.nome].adicionar_estoque(quantidade)

    def remover_produto(self, nome_produto, quantidade):
        if nome_produto in self.produtos:
            self.produtos[nome_produto].remover_estoque(quantidade)

    def exibir_estoque(self):
        for produto in self.produtos.values():
            print(produto)


# Exemplo de utilização do controle de estoque
if __name__ == "__main__":
    controle = ControleEstoque()

    # Adicionando produtos ao estoque
    controle.adicionar_produto(Produto("Camiseta", 100), 50)
    controle.adicionar_produto(Produto("Calça", 50), 20)
    controle.adicionar_produto(Produto("Sapato", 30), 10)

    # Exibindo estoque atual
    print("Estoque inicial:")
    controle.exibir_estoque()

    # Removendo produtos do estoque
    controle.remover_produto("Camiseta", 30)
    controle.remover_produto("Calça", 25)

    # Exibindo estoque após remoção
    print("\nEstoque após remoção:")
    controle.exibir_estoque()
```
